```python
# Paso a paso para crear un archivo .py en Visual Studio Code y agregarlo a un repositorio en GitHub

# 1. Abre Visual Studio Code.
# 2. Crea una nueva carpeta para tu proyecto (puedes hacerlo desde el explorador de archivos o desde VS Code).
# 3. Abre la carpeta en VS Code: Archivo > Abrir carpeta.
# 4. Dentro de la carpeta, crea un nuevo archivo: haz clic derecho en el explorador de archivos y selecciona "Nuevo archivo".
# 5. Nombra el archivo, por ejemplo: netflix_analysis.py
# 6. Copia y pega tu código Python en ese archivo:

import pandas as pd

df = pd.read_csv('netflix_data.csv')

year_movie = df[(df['release_year'] >= 1990) & (df['release_year'] < 2000)]

duration = df['duration'].mode().iloc[0]

print(duration)

short_movies_action = df[
    (df['release_year'] >= 1990) &
    (df['release_year'] <= 1999) &
    (df['genre'].str.lower() == 'action') &
    (df['duration'] < 90)
]

short_movie_count = short_movies_action.shape[0]

print(short_movie_count)

# 7. Guarda el archivo (Ctrl+S).

# 8. Abre una terminal en VS Code (Terminal > Nueva terminal).

# 9. Inicializa un repositorio Git en la carpeta del proyecto:
#    git init

# 10. Agrega el archivo al área de preparación (staging area):
#     git add netflix_analysis.py

# 11. Haz un commit con un mensaje descriptivo:
#     git commit -m "Primer análisis de datos de Netflix"

# 12. Crea un repositorio en GitHub (desde github.com, botón "New repository").

# 13. Copia la URL del repositorio (por ejemplo: https://github.com/tuusuario/tu-repo.git).

# 14. Enlaza tu repositorio local con el de GitHub:
#     git remote add origin https://github.com/tuusuario/tu-repo.git

# 15. Sube tus cambios a GitHub:
#     git push -u origin master

# ¡Listo! Ahora tu archivo .py está en tu repositorio de GitHub.
```
